```python
pie_data = pd.DataFrame(data.Disease.value_counts())
pie_data
```

```python
plt.figure(figsize = (20,10))
plt.subplot(121)
pie = plt.pie(pie_data.Disease, startangle = 90, counterclock = False, autopct='%1.1f%%')
plt.legend(pie_data.index)
plt.title("Pie Chart")
plt.axis("square")

plt.subplot(122)
bar = plt.bar(pie_data.index, pie_data.Disease.values)
plt.ylabel("Frequency")
plt.xlabel("Disease state")
plt.title("Bar Chart")

plt.suptitle("Distribution of Disease states", fontsize = 20)
```
